# PDF and CDF

```
import numpy as np
import matplotlib.pyplot as plt
```

```
from scipy.stats import norm
```

```
x = np.linspace(-6, 6, 1000)
```

```
fx = norm.pdf(x, loc=0, scale=1) # loc is mean and scale is stand. dev.
```

```
plt.plot(x, fx); # standard normal pdf.
```

```
Fx = norm.cdf(x, loc=0, scale=1)
```

```
plt.plot(x, Fx);
```

```
logfx = norm.logpdf(x, loc=0, scale=1)
```

```
plt.plot(x, logfx); # log pdf of the standard normal.
```

# Convolution

```
from PIL import Image
```

```
!wget https://media.gettyimages.com/photos/narrow-gauge-train-river-crossing-picture-id135328338?s=612x612
```

```
im = Image.open('narrow-gauge-train-river-crossing-picture-id135328338?s=612x612')
```

```
gray = np.mean(im, axis=2)
```

```
# create a 2-d gaussian filter.
x = np.linspace(-6, 6, 50)
```

```
fx = norm.pdf(x, loc=0, scale=1)
```

```
filt = np.outer(fx, fx)
```

```
plt.imshow(filt, cmap='gray')
```

```
from scipy.signal import convolve2d
```

```
out = convolve2d(gray, filt)
```

```
# plot images side by side using subplots.
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(out, cmap='gray')
```
