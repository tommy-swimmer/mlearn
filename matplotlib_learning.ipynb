# Line Chart

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0, 20, 1000)
```

```
y = np.sin(x) + 0.2 * x
```

```
# If you're not using a notebook
# Call plt.show()
plt.plot(x, y)
```

```
plt.plot(x, y)
plt.show()
```

```
# Some basic formatting
plt.plot(x, y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot'); # semicolon removes output 
```

# Scatterplot

```
X = np.random.randn(100, 2)
plt.scatter(X[:,0],X[:,1]) # first entry is horizontal axis.
```

* More specific data manipulation.

```
X = np.random.randn(200, 2)
# Want half of this data centered at some location, let's say 3,3.
X[:50] +=3 # add a 3 to all the elements up to 50.
Y = np.zeros(200)
Y[:50] = 1
# Draw scatterplot, use c (color) argument and pass y.
plt.scatter(X[:,0], X[:,1], c=Y)
```

```
# Plotting image.
plt.imshow(arr);
```

```
# Could use the original image as well.
plt.imshow(im)
```

```
# Turn into grayscale.
gray = arr.mean(axis=2)
```

```
gray.shape
```

```
plt.imshow(gray); # Why the weird colors?
# These are just randomly decided by matplotlib. as a heat-map.
```

```
# How to REALLY plot in grayscale.
plt.imshow(gray, cmap='gray')
```

# Exercise: Generate and Plot the XOR dataset.

```
# Make 1D arrays for x and y axes and spread them around -1 and 1
x1 = (np.random.random(2000)-0.5)*2
x2 = (np.random.random(2000)-0.5)*2

# Make an array for color picking
Y = np.zeros(2000)

for e in range(len(Y)):
  if (x1[e] <= 0 and x2[e] > 0) or (x1[e] > 0 and x2[e] <= 0):
    Y[e] = 1

# Plot it all
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Matplotlib Exercise')
plt.scatter(x1,x2,c=Y);

```
